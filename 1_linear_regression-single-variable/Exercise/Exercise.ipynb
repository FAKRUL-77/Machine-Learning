```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
x = df.drop('income', axis = 'columns')
y = df.drop('year', axis = 'columns')
x
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(x,y)
plt.plot(x,y)
```

```python
m_reg = linear_model.LinearRegression()
m_reg.fit(x,y)
```

```python
m = m_reg.coef_
m
```

```python
c = m_reg.intercept_
c
```

```python
x1 = 2018
y1 = m*x1+c
y1
```

```python
m_reg.predict([[2017]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(x,y)
plt.plot(x,m_reg.predict(x))
```

```python
df1 = pd.read_csv('year.csv')
df1
```

```python
income = m_reg.predict(df1)
df1['income'] = income
df1
```
